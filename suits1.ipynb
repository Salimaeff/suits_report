# Suits: A Synopsis

**_Suits_** is an American legal drama television series created and written by **Aaron Korsh**. The show premiered on USA Network in June 2011 and concluded in September 2019. It is set at a fictional law firm in New York City, and follows talented college dropout **Mike Ross**, who starts working as a law associate for **Harvey Specter**, despite never having attended law school. The show focuses on Harvey and Mike managing to close cases while maintaining Mike's secret.

## Main Characters

- **Harvey Specter** (played by **Gabriel Macht**): Known as one of the city's top litigators, Harvey is notorious for his high-flying style and technique in closing cases, partnered with his boundless ambition.
- **Mike Ross** (played by **Patrick J. Adams**): A brilliant associate blessed with a photographic memory, Mike’s lack of a legal degree is a constant threat to his and Harvey's secret.
- **Rachel Zane** (played by **Meghan Markle**): A paralegal who aspires to be an attorney but struggles with her test anxiety. She later becomes Mike's love interest.
- **Donna Paulsen** (played by **Sarah Rafferty**): Harvey's extremely competent executive assistant who provides a steadying influence at the firm. She knows all of Harvey's secrets and is fiercely loyal.
- **Louis Litt** (played by **Rick Hoffman**): A competitive but talented lawyer who often struggles with feelings of insecurity and jealousy towards Harvey's success.

## Themes and Reception

_Suits_ has been lauded for its witty dialogue, fast-paced storytelling, and strong character development. Although primarily a legal drama, the series infuses elements of comedy, making it a versatile and engaging show. The dynamic between Harvey and Mike, along with their personal struggles, forms the crux of the narrative, often overshadowed by the ongoing tension of maintaining Mike’s fraudulent secret.

Throughout its run, _Suits_ remained popular with viewers, offering an inside look into the legal battles, political plays, and personal relationships within a highly competitive Manhattan corporate law firm.

```python
import matplotlib.pyplot as plt

# Data
seasons = ["Season 1", "Season 2", "Season 3", "Season 4", "Season 5", "Season 6", "Season 7", "Season 8", "Season 9"]
viewers = [4.64, 4.48, 3.52, 2.76, 2.38, 1.92, 1.68, 1.30, 1.04]

# Graph of Viewership Over Time
plt.figure(figsize=(10, 5))
plt.plot(seasons, viewers, marker='o', color='b', linestyle='-')
plt.title('Viewership Over Time')
plt.xlabel('Season')
plt.ylabel('Average Viewership (millions)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Graph of Episode-to-Episode Changes in Viewership
viewers_diff = [0] + [viewers[i] - viewers[i-1] for i in range(1, len(viewers))]
plt.figure(figsize=(10, 5))
plt.bar(seasons, viewers_diff, color='g')
plt.title('Season-to-Season Changes in Viewership')
plt.xlabel('Season')
plt.ylabel('Change in Viewership (millions)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
